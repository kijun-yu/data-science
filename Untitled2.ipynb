```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.spatial import distance

# kmean Class
class kmean: 
    
    def __init__(self,dataf):# initialise data in class
        self.dataf = dataf
        self.kval = 0
        self.centeroids = list()
        self.cluster_id = [0]*len(dataf)
        
    def model(self): #recursive function to cluster data
        
        newcenteroids = [0]*self.kval #locally make new centroid
        for i in range(len(self.dataf)): #check each point's distance from centroid
            dist = 10000 # large value to initialise th longest distance
            for j in range(len(self.centeroids)):  # 1
                newdist = (distance.euclidean(self.centeroids[j],self.dataf[i])) #check the distance
                if newdist < dist: # given its closer than previous centeroid
                    self.cluster_id[i] = j #set the cluster id of point to this centroid
                    dist = newdist #set the distance to compare with next value
        #clustering is done with previous centroid
        # check the cluster to form newcentriod
        for i in range(self.kval): #number of clusters = kval
            datas = [self.dataf[idx] for idx, x in enumerate(self.cluster_id) if x == i] #select the only data with cluster id as present one
            newcenteroids[i] = [sum([item[0] for item in datas])/ len(datas),sum(item[1] for item in datas)/ len(datas)] #averrage out x value and y value to find new centeriods
            
        if newcenteroids == self.centeroids: #if there is no change made for centroid finish recursion
            return
        else: #if there is change
            self.centeroids = newcenteroids #set the centroid as newly created one
            self.model() #recurse through same function to form cluster again
            return
            
    def clusterise(self, kval): #make clusters with given data (kmeans algorithm)
        self.kval = kval
        self.centeroids = list() #initialise the centroids
        for i in range(kval):
            index = np.random.randint(0,len(self.dataf))#randomly pick number between 0 to num of data
            self.centeroids.append(self.dataf[index]) #set one point as centroid
        self.model() #after selecting centroids, start clustering     
    
with open("dataset.txt",'r') as f:
    dataset = [line.split() for line in f]

points = list()
for i in range(len(dataset)):
    l = [word.split(',')for word in dataset[i]]
    points.extend(l)
for i in points:
    i[0],i[1] = float(i[0]),float(i[1])
    
kval = 5

kmeans = kmean(points)
kmeans.clusterise(kval)

df = pd.DataFrame(data = kmeans.dataf, columns=('x','y'))


plt.figure(figsize = (16,11))
plt.scatter(df['x'], df['y'],s = 8, c = kmeans.cluster_id)
plt.show()
```
