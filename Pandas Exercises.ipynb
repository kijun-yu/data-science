```python
l = [
    {
        "Student":[{"Exam":90,"Grade":"A"},
                  {"Exam":99,"Grade":"B"},
                  {"Exam":95,"Grade":"C"}],
        "Name": "Ian"
    },
    {
        "Student":[{"Exam":90,"Grade":"A"},
                  {"Exam":85,"Grade":"B"}],
        "Name": "Jacob"
    }
]

print(l)
```

```python
rows =[]

for data in l:
    data_row = data["Student"]
    n = data["Name"]
    
    for row in data_row:
        row["Name"] = n
        rows.append(row)
        
df = pd.DataFrame(rows)
```

```python
df
```

```python
df = df.pivot_table(index = "Name", columns = ["Grade"], values= ["Exam"]).reset_index()
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
#1. divide it into 3 groups (setosa, virginica, versicolor)
#2. from each group. sort them by any of 4 column
#3. find average of each column from 3 groups
#4. bring which column shows dramatic diffrence(largest abs value in difference)
df
```

```python
#1. divide it into 3 groups (setosa, virginica, versicolor)
setosa = df[df["species"] == "setosa"]
virginica = df[df["species"] == "virginica"]
versicolor = df[df["species"] == "versicolor"]
```

```python
versicolor
```

```python
#2. from each group. sort them by any of 4 column
sort_setosa = setosa.sort_values(by = "petal_width")
sort_virginica = virginica.sort_values(by = "petal_width")
sort_versicolor = versicolor.sort_values(by = "petal_width")
```
