```python
y1 = np.random.normal(size = 50).cumsum()

# (left) default xlim ylim
plt.subplot(121)
plt.plot(y1)
plt.title("Y1")

# (Right) using custom xlim ylim
plt.subplot(122)
plt.plot(y1)
plt.xlim(0,100)
plt.ylim(-10,10)
plt.title("Y2")

plt.show()
```

```python
x = np.linspace(0,2 * np.pi, 20)

# blue, square
plt.plot(x,np.sin(x), "bs")

# green, --, triangle points
plt.plot(x,np.cos(x), "g--^")

# red, dotted line, o points
plt.plot(x,np.sin(2*x), "r:o")

plt.title ("Various line plots")
plt.legend(["bs","g--^","r:o"])
plt.show()
```

```python
x = np.linspace(0,2* np.pi, 10)
y = np.sin(x)

xerr = np.linspace(0,1,10) / 5
yerr = np.linspace(1,0,10) / 5

plt.errorbar(x,y,yerr,xerr)

plt.title("Errorbars")
plt.legend(["Error"])

plt.show()
```

```python
x = np.linspace(0,2* np.pi, 10)
y = np.sin(x)
xerr = np.linspace(0,1,10) / 5
yerr = np.linspace(1,0,10) / 5


plt.errorbar(x,y,yerr,xerr, label = "both limits")

plt.errorbar(x,y + 1,yerr = yerr, uplims = True, label = "uplims = True")

plt.errorbar(x,y + 2,yerr = yerr, uplims = [True,False] * 5, lolims = [False, True] * 5,label = "vice versa")

plt.title("Errorbars")
plt.legend()


plt.show()
```
